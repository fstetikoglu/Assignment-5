```python
import glob
import numpy 
import matplotlib
import pandas as pd
import matplotlib.pyplot 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

dataset = sorted(glob.glob('data*.csv'))
print(dataset)
```

```python
def visualize(filename):
    data = pd.read_csv(filepath_or_buffer = filename, header=None)
    minValues = data.min()
    maxValues = data.max()
    avrValues = data.mean()
    columns =[i for i in range(40)]
    plt.plot(columns , minValues, label='min')
    plt.plot(columns , maxValues, label='max')
    plt.plot(columns , avrValues, label='avg')
    plt.ylabel('Values')
    plt.xlabel('Feature#s')
    plt.legend()
    plt.show()
```

```python
def identify(filename):
    data = pd.read_csv(filepath_or_buffer = filename, header=None)
        
    if data.max()[0]==0 and data.max()[20]==20:
        print ("I doubt the integrity of this dataset.")
    elif all(j==0 for j in data.min()):
        print ('Why is this happening?')
    else:
        print ('Dataset looks good')
```

```python
for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    visualize(datasets)
    identify(datasets)
```
