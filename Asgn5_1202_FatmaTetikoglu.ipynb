```python
#import libraries
import glob
import numpy 
import matplotlib
import pandas as pd
import matplotlib.pyplot 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

#define and load dataset(s)
dataset = sorted(glob.glob('data*.csv'))
print(dataset)
```

```python
#define the new function for visuals
def visualize(filename):
    """Plot the graphs for min, max and average"""
    data = pd.read_csv(filepath_or_buffer = filename, header=None) #load the datasets
    
    minValues = data.min()  #finds the min values of each column of each datasets
    maxValues = data.max()  #finds the min values of each column of each datasets
    avrValues = data.mean()  #finds the min values of each column of each datasets
    columns =[i for i in range(40)]  #rename the features by a list
    
    plt.plot(columns , minValues, label='min')  #plot the line graph for min values of each datasets
    plt.plot(columns , maxValues, label='max')  #plot the line graph for max values of each datasets
    plt.plot(columns , avrValues, label='avg')  #plot the line graph for average values of each datasets
    
    plt.ylabel('Values')  #label the y-axis
    plt.xlabel('Feature#s')   #label the x-axis
        
    plt.legend()   #assigne the legends
     
    plt.show()   #show the graph
```

```python
#define a new funtion to identify issues
def identify(filename):
    """identify the issues with dataset"""
    data = pd.read_csv(filepath_or_buffer = filename, header=None)  #load the dataset
        
    if data.max()[0]==0 and data.max()[20]==20:     #check whether min of features#0s=0 and max of feature#20=20
        print ("I doubt the integrity of this dataset.")
    elif sum(data.min())==0:       #check whether the sum of min of all feature=0
        print ('Why is this happening?')
    else:
        print ('Dataset looks good.')
```

```python
#operate the functions defined above
for datasets in dataset[:4]:
    print("\n")
    print(datasets)
    plt.title('Dataset: {}'.format(datasets))
    visualize(datasets)
    identify(datasets)
```
